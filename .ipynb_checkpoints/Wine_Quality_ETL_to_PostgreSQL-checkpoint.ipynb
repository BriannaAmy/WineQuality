```python
import pandas as pd
from pprint import pprint
```

```python
#Read in CSV files
white_wine = pd.read_csv("winequality-white.csv", sep=";")
red_wine = pd.read_csv("winequality-red.csv", sep=";")
```

```python
#get the index type and column names
white_wine.info()
red_wine.info()
```

```python
#put the columns into a list
white_wineCols = white_wine.columns.tolist()
white_wineCols

red_wineCols = red_wine.columns.tolist()
red_wineCols
```

```python
#get each value into a list
white_winevalues = white_wine.values.tolist()
white_winevalues

red_winevalues = red_wine.values.tolist()
red_winevalues
```

```python
#creating a data frame for white_wine
white_wineDF = pd.DataFrame(white_winevalues, columns = white_wineCols)
white_wineDF
```

```python
red_wineDF = pd.DataFrame(red_winevalues, columns = red_wineCols)
red_wineDF
```

```python
#adding a "Type" column to both dataframes. 0 represents White Wine and 1 represents Red Wine.
white_wineDF["Type"] = 0
red_wineDF["Type"] = 1
```

```python
#check to make sure the "Type" for white_wine is showing "0"
white_wineDF.head()
```

```python
#check to make sure the "Type" for red_wine is showing "1"
red_wineDF.head()
```

```python
#group white_wine quality into 3 different groups. (0 - Bad, 1 - Average, 2 - Great)
def categorize_quality(quality):
    if 1 <= quality <= 3:
        return '0'
    elif 4 <= quality <= 6:
        return '1'
    elif 7 <= quality <= 9:
        return '2'

#create the 'quality_category' column
white_wineDF['quality_category'] = white_wineDF['quality'].apply(categorize_quality)

#check to make sure the quality_category column is added
white_wineDF.head()
```

```python
#group red_wine quality into 3 different groups. (0 - Bad, 1 - Average, 2 - Great)
def categorize_quality(quality):
    if 1 <= quality <= 3:
        return '0'
    elif 4 <= quality <= 6:
        return '1'
    elif 7 <= quality <= 9:
        return '2'

#create the 'quality_category' column
red_wineDF['quality_category'] = red_wineDF['quality'].apply(categorize_quality)

#check to make sure the Quality_category column is added
red_wineDF.head()
```

```python
#create a third dataframe which merges white_wine and red_wine
combinedDF = pd.concat([white_wineDF, red_wineDF])
combinedDF
```

```python
#view summary statistics for the red_wineDF DataFrame
red_wineDF.describe()
```

```python
#add an ID column that serves as a primary key
white_wineDF['id'] = white_wineDF.index
```

```python
red_wineDF['id'] = red_wineDF.index
```

```python
combinedDF['id'] = combinedDF.index
```

```python
#view the DataFrame to make sure an ID column was added
white_wineDF.head()
```

```python
#LOADING INTO POSTGRE SQL
from sqlalchemy import create_engine
from sqlalchemy import Column, Integer, String, Float
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import DeclarativeBase
from sqlalchemy import DateTime

class Base(DeclarativeBase):
    pass
```

```python
# Create Database Connection for Postgres
dbName = 'WineQuality'
userName = 'postgres'
password = 'password'
engine = create_engine(f"postgresql://{userName}:{password}@localhost/{dbName}")
conn = engine.connect()
```

```python
from sqlalchemy.orm import Session 
session = Session(bind=engine)
```

```python
white_wineDF.info()
```

```python
class Combined(Base):
    __tablename__ = "Combined (White and Red)"
    id = Column(Integer, primary_key=True)
    fixed_acidity = Column(Float)
    volatile_acidity = Column(Float)
    citric_acid = Column(Float)
    residual_sugar = Column(Float)
    chlorides = Column(Float)
    free_sulfur_dioxide = Column(Float)
    total_sulfur_dioxide = Column(Float)
    density = Column(Float)
    pH = Column(Float)
    sulphates = Column(Float)
    alcohol = Column(Float)
    quality = Column(Integer)
    quality_category = Column(String)
    type = Column(Integer)
```

```python
Base.metadata.create_all(engine)
```

```python
combinedDF.to_sql('WineQuality', engine, if_exists='replace', index=False)
```

```python
session.commit()
```

```python
# Define the SQL query
query = 'SELECT * FROM "WineQuality"'

try:
    # Execute the query and create a DataFrame
    df = pd.read_sql_query(query, engine)
    # Display the DataFrame
    print(df.head())
except Exception as e:
    print(f"Error: {e}")
finally:
    # Close the connection
    engine.dispose()
```

```python
# Split the DataFrame
WhiteWine = df[df['Type'] == 0]
RedWine = df[df['Type'] == 1]
```
